```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Reload the CSV file
df_new = pd.read_csv('D:\\plotBrems\\plotBrems\\plotBrems\\bin\\Debug\\log.csv', sep=';', header=None, names=["Column1", "Column2"])

y_ticks = [0, df_new['Column1'].str.replace(',', '.').astype(float).max(), df_new['Column1'].str.replace(',', '.').astype(float).min()]

# Convert the 'Column1' data to the data type int64
df_new['Column1'] = df_new['Column1'].str.replace(',', '.').astype(float)

# Plotting the first column
plt.figure(figsize=(10, 6))
plt.plot(df_new['Column1'], marker='o')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Values from the First Column of log.csv')
plt.yticks(y_ticks, y_ticks)  # Set y-axis ticks
plt.grid(True)
plt.show()
```
